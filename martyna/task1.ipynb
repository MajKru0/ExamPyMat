```python
import numpy as np
import matplotlib.pyplot as plt
import rasterio
```

```python
with rasterio.open('C:/Users/User/Desktop/egzamin/t1_lst2023_Jul_Aug.tif') as src:
    temp1 = src.read(1)

with rasterio.open('C:/Users/User/Desktop/egzamin/t1_ndvi2023_Jul_Aug.tif') as src:
    ndvi1 = src.read(1)

with rasterio.open('C:/Users/User/Desktop/egzamin/t1_lst2024May.tif') as src:
    temp2 = src.read(1)

with rasterio.open('C:/Users/User/Desktop/egzamin/t1_ndvi2024May.tif') as src:
    ndvi2 = src.read(1)
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(temp1, cmap='hot')  
plt.colorbar()
plt.title('Temperature Image 2023')

plt.subplot(2, 2, 2)
plt.imshow(ndvi1, cmap='jet')  
plt.colorbar()
plt.title('NDVI Image 2023')
```

```python
plt.subplot(2, 2, 3)
plt.hist(ndvi1.ravel(), bins=50, color='blue', alpha=0.7)
plt.title('NDVI Histogram 2023')
plt.xlabel('NDVI Values')

plt.subplot(2, 2, 4)
plt.hist(temp1.ravel(), bins=50, color='red', alpha=0.7)
plt.title('Temperature Histogram 2023')
plt.xlabel('Temperature Values')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(ndvi1.ravel(), temp1.ravel(), s=5, alpha=0.5, c='purple')
plt.xlabel('NDVI')
plt.ylabel('Temperature')
plt.title('Scatter Plot: NDVI vs Temperature 2023')
plt.grid(True)
plt.show()
```
